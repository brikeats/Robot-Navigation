```python

from scipy.interpolate import interp1d
import time
import numpy as np
import matplotlib.pyplot as plt


"""
Treatment here: http://planning.cs.uiuc.edu/node659.html

Units are cm, sec, radians.

## Robot parameters ##

# geometry
robot_width: distance between wheel planes (assumed parallel!)
wheel_radius: in cm

# sensor sweep
phi_range: the range of the sensor servo is -phi_range/2...phi_range/2
phi_n: number of steps to go from -phi_range/2...phi_range/2
phi_T: time to go from -phi_range/2...phi_range/2

## Robot state ##
pos: (x,y) coordinates in room frame
theta: angle of heading in coordinate frame (where heading is perpendicular to axis)
ur, ul: the speed of the left/right motors, assumed perfectly controllable


"""


def logistic_curve(x, k=1):
    return 1./(1+np.exp(-k*x))


def intersect(line1, line2):
    x1, x2, y1, y2 = line1
    x3, x4, y3, y4 = line2
    den = (x1-x2)*(y3-y4) - (y1-y2)*(x3-x4)
    if den == 0:  # parallel
        return None
    px = (x1*y2 - y1*x2) *(x3-x4) - (x1-x2)*(x3*y4 - y3*x4)
    py = (x1*y2 - y1*x2) *(y3-y4) - (y1-y2)*(x3*y4 - y3*x4)
    return np.array([px/den, py/den])


def intersect_walls(pt1, pt2, walls):
    line = [pt1[0], pt2[0], pt1[1], pt2[1]]
    intersect_pts = []
    for wall in walls:
        intersect_pt = intersect(line, wall)
        if intersect_pt is not None:  # not parallel; intersection exists
            vec1 = pt2 - pt1
            vec2 = intersect_pt - pt1
            dot = np.dot(vec1, vec2)
            if dot > 0:  # pointing in right direction (forward)
                wall_vec = [wall[0]-wall[1], wall[2]-wall[3]]
                vec1 = intersect_pt - [wall[0], wall[2]]
                vec2 = intersect_pt - [wall[1], wall[3]]
                dot1 = np.dot(wall_vec, vec1)
                dot2 = np.dot(wall_vec, vec2)
                if dot1*dot2 <= 0:  # it is on the wall
                    vec = pt1 - pt2
                    vec1 = intersect_pt - pt1
                    vec2 = intersect_pt - pt2
                    dot1 = np.dot(vec, vec1)
                    dot2 = np.dot(vec, vec2)
                    if dot1*dot2 <= 0:  # it is between pt1 and pt2
                        intersect_pts.append(intersect_pt)
    dist_vecs = [pt1-pt for pt in intersect_pts]
    dists = [np.sqrt(vec[0]**2+vec[1]**2) for vec in dist_vecs]
    try:
        min_ind = np.argmin(dists)
        return intersect_pts[np.argmin(dists)]
    except ValueError:
        pass


def draw_lines(lines, plots=None, colors=None, **kwargs):
    if colors is None:
        colors = ['k' for _ in range(len(lines))]
    if plots is None:
        line_plots = []
        for color, (ista, iend, jsta, jend) in zip(colors, lines):
            line_plot, = plt.plot([ista, iend], [jsta, jend], color=color, **kwargs)
            line_plots.append(line_plot)
        return line_plots
    else:
        for plot, (ista, iend, jsta, jend) in zip(plots, lines):
            plot.set_data([ista, iend], [jsta, jend])


            
class RangeSensor:
    
    def __init__(self, walls, noise):
        self.walls = walls
        self.noise = noise
        self.line_plot = None
        
    ### Get range readings
    def sense(self, pos, theta):
        readings = {}
        for phi in np.linspace(theta-phi_range/2, theta+phi_range/2, phi_n):
            ds = sensor_range*np.array([np.cos(phi), np.sin(phi)])
            intersect_pt = intersect_walls(pos, pos+ds, self.walls)
            # TODO: restrict to phi_range
            vec = pos - intersect_pt
            dist = np.sqrt(vec[0]**2 + vec[1]**2)
            dist += np.random.rand()
            # the range readings are associated with the servo angle
            readings[phi-theta] = dist + noise*np.random.randn()
        return readings
    
    def draw_lines(self, readings, pos, theta):
        lines = []
        for phi, reading in readings.iteritems():
            dx = reading*np.cos(theta+phi)
            dy = reading*np.sin(theta+phi)
            line = np.array([pos[0], pos[0]+dx, pos[1], pos[1]+dy])
            lines.append(line)
        if self.line_plot is None:
            self.line_plot = draw_lines(lines)
        else:
            draw_lines(lines, self.line_plot)

```

```python

def check_for_collisions(pos, theta, walls):
    n = np.array([np.cos(theta-np.pi/2), np.sin(theta-np.pi/2)])  # heading
    L = pos + robot_width*n/2
    R = pos - robot_width*n/2
    B = pos - robot_length*np.array([np.cos(theta), np.sin(theta)])

    if intersect_walls(L, R, walls) is not None:
        print 'front collision!'
        return True
    elif intersect_walls(L, B, walls) is not None:
        print 'left collision!'
        return True
    elif intersect_walls(R, B, walls) is not None:
        print 'right collision!'
        return True
    return False


class Robot:
    
    def __init__(self, length, width, wheel_radius):
        self.length = length
        self.width = width
        self.wheel_radius = wheel_radius
        self.line_plot = None
        self.uL, self.uR = 0., 0.
        
        self.start_turning_thresh = 0.03
        self.stop_turning_thresh = 5*self.start_turning_thresh
        self.force_ewma = self.stop_turning_thresh  # don't start at 0 or it will execute turnabout immediately
        self.alpha = 0.9
        self.turning_mode = False
        
        
    def draw(self, pos, theta, plots=None, **kwargs):
        n = np.array([np.cos(theta-np.pi/2), np.sin(theta-np.pi/2)])  # heading
        L = pos + self.width*n/2
        R = pos - self.width*n/2
        h = np.array([np.cos(theta), np.sin(theta)])
        B = pos - self.length*h
        front_line = np.array([L[0], R[0], L[1], R[1]])
        left_line = np.array([L[0], B[0], L[1], B[1]])
        right_line = np.array([R[0], B[0], R[1], B[1]])
        lines = [front_line, left_line, right_line]
    
        robot_colors = ['r','b','b']
        if self.line_plot is None:
            self.line_plot = draw_lines(lines, colors=robot_colors, **kwargs)
        else:
            draw_lines(lines, self.line_plot, colors=robot_colors, **kwargs)
    
    def readings_to_force(self, readings, stopping_distance=30, steepness=0.1):
        net_force = np.array([0., 0.])
        for phi, reading in readings.iteritems():
            force = 2*(logistic_curve(reading-stopping_distance, steepness)-0.5)
            net_force += force * np.array([np.cos(phi), np.sin(phi)])
        net_force /= float(len(readings))
        return net_force
    
    def turn_mode_is_finished(self, readings):
        force = self.readings_to_force(readings)
        f = force[0]**2 + force[1]**2
        if f < self.stop_turning_thresh:
            return False
        else:
            return True

    def update_control_effort(self, readings, lambda_=10, omega0=10, scale=10):
        
        if self.turning_mode:
            net_force = np.array([0, -0.2])
            if self.turn_mode_is_finished(readings):
                self.turning_mode = False
        else:
            net_force = self.readings_to_force(readings)
        
        # speed in robot frame
        self.force_ewma = (1-alpha)*net_force + alpha*self.force_ewma
        f_ewma = self.force_ewma[0]**2 + self.force_ewma[1]**2
        f = net_force[0]**2 + net_force[1]**2
        if f_ewma < self.start_turning_thresh:
            self.turning_mode = True

        net_force *= float(scale)     
        v = net_force[0]
        omega = net_force[1] + omega0/(np.sqrt(np.sum(net_force**2)))
        self.uR = (v + lambda_*omega)/(lambda_+1)
        self.uL = (v - lambda_*omega)/(lambda_+1)
    


### robot parameters
robot_width = 29.  # distance b/t wheels
robot_length = 17.5  # distance b/t caster and front axel
wheel_radius = 4.5
robot = Robot(robot_length, robot_width, wheel_radius)

noise = 3.
sensor = RangeSensor(walls, noise)

phi_range = np.pi/2
phi_n = 5
phi_T = 2.
sensor_range = 400.


### initial robot state
pos = np.array([sz/2, 3*sz/2], dtype=float)  # position (center of wheel axis)
theta = np.pi
ul, ur = 0.5, 0.5  # wheel speed in [rad/s]
lambda_ = 35
stopping_distance = 20
steepness = 0.1


%pylab
# %matplotlib inline


# create initial plots
wall_plot = draw_lines(walls, linewidth=3)
robot.draw(pos, theta, linewidth=3)
readings = sensor.sense(pos, theta)
# sensor.draw_lines(readings, pos, theta)
plt.axis('equal')



dt = .02

for t in range(1, 2000):

    robot_hit_wall = check_for_collisions(pos, theta, walls)
    if robot_hit_wall:
        break
    
    robot.draw(pos, theta, linewidth=3)
    readings = sensor.sense(pos, theta)
    sensor.draw_lines(readings, pos, theta)
    
    robot.update_control_effort(readings, lambda_=10, omega0=100, scale=70)
    ul, ur = robot.uL, robot.uR

    ### velocity in global frame
    vx = wheel_radius * (ul+ur) * np.cos(theta)
    vy = wheel_radius * (ul+ur) * np.sin(theta)
    omega = wheel_radius / robot_width * (ur-ul)

    ### propagate in time
    pos += np.array([vx*dt, vy*dt])
    theta += omega*dt
    
    
    plt.title('t = %i' % t)
    plt.draw()
#     time.sleep(0.1)
```
